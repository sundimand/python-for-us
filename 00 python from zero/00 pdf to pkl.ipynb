```python
import os
import pandas as pd
import fitz  # pip install pymupdf
import re

# Fungsi pembersih teks
def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9.,;:()\s]', '', text)  # hanya karakter umum
    text = re.sub(r'\s+', ' ', text)                  # gabungkan spasi berlebih
    return text.strip()

# Path ke folder PDF
folder_path = "dataarticle"  # ganti dengan path folder kamu

# Ambil semua file PDF
pdf_files = [f for f in os.listdir(folder_path) if f.lower().endswith(".pdf")]

# Proses semua file dan simpan ke DataFrame
data = []
for file_name in pdf_files:
    file_path = os.path.join(folder_path, file_name)
    #print(f"Memproses: {file_name}")
    try:
        doc = fitz.open(file_path)
        text = ""
        for page in doc:
            page_text = page.get_text("text")  # atau "blocks" untuk layout-aware
            text += page_text + " "
        cleaned = clean_text(text)
        data.append({"articlename": file_name, "text": cleaned})
    except Exception as e:
        print(f"Gagal ekstraksi: {file_name} â†’ {e}")
        data.append({"articlename": file_name, "text": ""})

# Simpan ke DataFrame
df = pd.DataFrame(data)
df = df.rename(columns={'text': 'textdata'})

# Tampilkan hasil
#print(df.head())
```

df to pickle

```python
df.to_pickle(f"{folder_path}.pkl")
```
